Put the image generated by fm-based model with MindEyeV2

```python
import torch
import matplotlib.pyplot as plt
from torchvision import transforms

# Load the enhanced reconstructions
model_name = "debug_finetune_smallModel_subsetSessions"  # or whatever your model name is
all_enhancedrecons = torch.load(f"evals/{model_name}/{model_name}_all_enhancedrecons.pt")

print(f"Loaded enhanced reconstructions shape: {all_enhancedrecons.shape}")

# Display the first image (which would be shown when plotting=True before the err)
img_idx = 0  # First image
enhanced_recon = all_enhancedrecons[img_idx]

# Convert to PIL and display (same as the script does)
plt.figure(figsize=(8, 8))
plt.imshow(transforms.ToPILImage()(enhanced_recon))
plt.title(f"Enhanced Reconstruction - Image {img_idx}")
plt.axis('off')
plt.show()

# If you want to see multiple images
num_to_show = min(5, len(all_enhancedrecons))  # Show first 5 or all if less than 5
fig, axes = plt.subplots(1, num_to_show, figsize=(4*num_to_show, 4))
if num_to_show == 1:
    axes = [axes]

for i in range(num_to_show):
    axes[i].imshow(transforms.ToPILImage()(all_enhancedrecons[i]))
    axes[i].set_title(f"Enhanced Recon {i}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
